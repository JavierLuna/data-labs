```python
# Voy a borrar las columnas Unnamed 22 y Unnamed 23, para ello empezaré haciendo una copia de mi data frame.
df_aux = df.copy()

#Y ahora borro las columnas.

df_aux.drop(columns=['Unnamed: 22','Unnamed: 23'],inplace=True)

```

```python
#Creo una columna para ver si las columnas href formula y href son iguales y saco el numero de filas que son iguales (5989, por lo que tengo 3 filas diferentes)
df_aux.head()
df_aux['pdf_equals'] = (df_aux['href formula'] == df_aux['href'])
df_aux['pdf_equals'].sum()

```
