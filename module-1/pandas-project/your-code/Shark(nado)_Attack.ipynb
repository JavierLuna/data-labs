Borranmos las columnas Unnamed 22 y Unnamed 23 (por tener prácticamente todos los valores en null) para ello empezaré haciendo una copia del data frame.

```python
df_aux = df.copy()
```

Una vez tenemos la copia, borramos la columna.

```python
df_aux.drop(columns=['Unnamed: 22','Unnamed: 23'],inplace=True)
```

Vemos como ya no tenemos las columnas que hemos borrado.

```python
df_aux.head(5)
```

Creamos una columna para ver si las columnas href formula y href son iguales y sumamos el numero de filas que son iguales en las dos columnas (5989, por lo que tenemos 54 filas diferentes).

Creamos una columna para ver si las columnas Case Number.1 y Case Number.2 son iguales y sumamos el numero de filas que son iguales en las dos columnas (5981, por lo que tengo 11 filas diferentes).

```python
df_aux.head()
df_aux['CaseNumber_equals'] = (df_aux['Case Number.1'] == df_aux['Case Number.2'])
df_aux['CaseNumber_equals'].sum()
```

Hacemos lo mismo con la Columna Case Number

```python
df_aux['CaseNumber_equals3'] = (df_aux['Case Number.2'] == df_aux['Case Number'])
df_aux['CaseNumber_equals3'].sum()
```

```python
df_aux['CaseNumber_equals2'] = (df_aux['Case Number.1'] == df_aux['Case Number'])
df_aux['CaseNumber_equals2'].sum()
```

Borramos las columnas innecesarias.

```python
df_aux.drop(columns=['Case Number.1','Case Number.2','CaseNumber_equals2','CaseNumber_equals3','CaseNumber_equals','href_equals','href formula'],inplace=True)
```

Miramos los nombres de las columnas para ver si alguna está escrita de manera diferente.

```python
df_aux.columns
```

Vamos a cambiar el nombre de la columna 'Sex_' por 'Sex'.

```python
df_aux.rename(columns = {'Sex ':'Sex'},inplace = True)
```

Miramos los distintos tipos de valores para columna Sex.

```python
df_aux.Sex.value_counts()
```

Remplazamos los valores por Male y Female.

```python
df_aux.replace('M','Male',inplace = True)
df_aux.replace('F','Female',inplace = True)
df_aux.replace('N','-',inplace = True)
df_aux.replace('lli','-',inplace = True)
df_aux.replace('.','-',inplace = True)
df_aux.replace('M ','Male',inplace = True)
```

Volvemos a mirar los valores para ver que se ha registrado el cambio.

```python
df_aux.Sex.value_counts()
```

Remplazamos todos los valores nulos por un -.

```python
df_aux.fillna('-', inplace = True)
```

Marcamos la columna 'original order' como indice.

```python
df_aux.set_index('original order',inplace = True)
```

Quitamos los años que son menores de 1000.

```python
df_aux.drop(df_aux[df_aux.Year < 1000].index, inplace = True)
```

Remplazamos los Valores de la columna Fatal y le cambiamos el nombre.

```python
df_aux['Fatal (Y/N)'].value_counts()
```

```python
df_aux['Fatal (Y/N)'].replace('UNKNOWN','False', inplace = True)
df_aux['Fatal (Y/N)'].replace('-','False', inplace = True)
df_aux['Fatal (Y/N)'].replace(' N','False', inplace = True)
df_aux['Fatal (Y/N)'].replace('N ','False', inplace = True)
df_aux['Fatal (Y/N)'].replace('Female','False', inplace = True)
df_aux['Fatal (Y/N)'].replace('#VALUE!','False', inplace = True)
df_aux['Fatal (Y/N)'].replace('n','False', inplace = True)
df_aux['Fatal (Y/N)'].replace('Y','True', inplace = True)
```

```python
df_aux['Fatal (Y/N)'].value_counts()
```

Cambiamos el tipo de Activity a Categoría.

```python
df_aux['Activity'] = df_aux.Activity.astype('category')
```

```python
df_aux.dtypes
```

Filtramos por los ataques de tiburón en España

```python
df_aux_filtered= df_aux[df_aux['Country']== 'SPAIN' ]
```
