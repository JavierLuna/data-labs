```python
import pandas as pd
import numpy as np
```

```python
#Import the data using Pandas: Its a csv
#No encortraba el archivo, por lo que lo he pegado dentro de your code --> preguntar cómo se generan las rutas de los archivos.

df = pd.read_csv('GSAF5.csv', header=0, encoding = 'unicode_escape')
```

```python
#Echo un vistazo rápido a las primeras filas del df
df.head(8)
```

```python
#Vemos el tamaño y estructura del dataset con numpys
np.size(df)
np.shape(df)
np.sum(df)
np.mean(df)
```

```python
#Vemos que hay columnas con valores null, vamos a revisar en cuales y cuántos hay por columna
null_cols = df.isnull().sum()
null_cols[null_cols>0]
```

```python
#Las columnas Unnamed: 22 y Unnamed: 23 no tienen casi info. Vamos a eliminar aqullas columnas con más de 4000 nulls
drop_cols = list(null_cols[null_cols > 4000].index)
#Reviso cuáles son
print(df.columns, drop_cols)
#Las elimino
df = df.drop(drop_cols, axis=1)
```

```python
#Reviso que esté bien(con otro método)
df.isnull().sum()
```

```python
#Revisamos los tipos de datos
df.dtypes
```

```python
#Cambio el tipo de dato de year a objeto y reviso
df['Year'] = df['Year'].astype('object')
#Intento cambiar el date pero da error 
#df['Date'] = df['Date'].astype('datetime')
df.dtypes
```

```python
#Cambio / por - en Date
df['Date'] = df['Date'].str.replace('/', '-')
df.Date
```

```python
#Voy a quitar duplicados y revisar cuántos he quitado
before = len(df)
data = df.drop_duplicates()
after = len(df)
print('Number of duplicate records dropped: ', str(before - after))
```

```python
# Reviso si'Case Number.1' y 'Case Number.2' son iguales
num_iguales = 0
for i in df.index:
    if df['Case Number.1'][i] == df['Case Number.2'][i]:
        num_iguales += 1
print(np.shape(df),num_iguales)
```

```python
#Son prácticamente iguales, por lo que elimino una de ellas
df.drop(['Case Number.2'], axis=1, inplace = True)
```
