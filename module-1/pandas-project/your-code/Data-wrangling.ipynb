```python
# Observamos que hay especies que no tienen registro, sin embargo, hay filas que no tienen especie pero son fatales 
# Por tanto, vamos a tener como condicionante máximo que se eliminarán registro que no sean representativos 
# siempre que no correspondan a ataques fatales 
        
ataques = ataques.drop(ataques[(ataques['Species '].isnull()) & (ataques['Fatal (Y/N)'] == 'N')].index)
```

```python
# Renombramos columnas 
ataques.rename(index=str, columns={ 'Sex'  :  'Sex' ,  'Species'  : 'Species','Fatal (Y/N)':'Fatal(Y/N)'},inplace=True) 
```

```python
#Replace no modifica el dataframe. Hay que poner inplace para que lo modifique

ataques['Fatal(Y/N)'] = ataques['Fatal(Y/N)'].replace("Y",True).replace("F", True, inplace = True)
ataques['Fatal(Y/N)'].value_counts()

#hacer una funcion que si no es True devuelva False
```

```python
# Eliminamos espacios vacíos e innecesarios en los campos de tipo string 
ataques['Type'] =ataques['Type'].str.strip() 
ataques['Country'] =ataques['Country'].str.strip() 
ataques['Area'] =ataques['Area'].str.strip()     
ataques['Location'] =ataques['Location'].str.strip() 
ataques['Activity'] =ataques['Activity'].str.strip() 
ataques['Name'] =ataques['Name'].str.strip() 
ataques['Injury'] =ataques['Injury'].str.strip()  
ataques['Investigator or Source'] =ataques['Investigator or Source'].str.strip() 

#lower.header(ataques)
```
