```python
import pandas as pd
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
def get_csv(csv_file, enconding_value):
    df = pd.read_csv(csv_file, encoding = enconding_value, na_values=["-", False])
    return df

superheroes = get_csv("superheroes.csv", "latin-1")
superheroes_powers= get_csv("superheroespowers.csv", "latin-1")
```

```python
def rename_column(df, old_column, new_column):
    return df.rename(columns = {old_column: new_column}, inplace = True)

superherores = rename_column(superheroes, "name", "Name")

superheroes.head()
```

```python
def capitalize_column_values(df, column):
    df[column] = df[column].str.capitalize()

capitalize_column_values(superheroes, "Alignment")
capitalize_column_values(superheroes, "Eye color")
capitalize_column_values(superheroes, "Skin color")

superheroes.head()
```

```python
def pounds_to_kilograms(df, column):
      df[column] = df[column].apply(lambda value: (value/2.205) if value > 0 else None)
            
pounds_to_kilograms(superheroes, "Weight")

superheroes.head()
```

```python
def feet_to_meters(df, column):
    df[column] = df[column].apply(lambda value: (value/3.281) if value > 0 else None)
            
feet_to_meters(superheroes, "Height")

superheroes.head()
```

```python
def rename_column(df, old_column, new_column):
    return df.rename(columns = {old_column: new_column}, inplace = True)

superherores = rename_column(superheroes, "Height", "Height (m)")
superherores = rename_column(superheroes, "Weight", "Weight (kg)")

superheroes.head()
```

```python
def change_columns_order(df, list_new_columns_order):
    return df[list_new_columns_order]

superheroes = change_columns_order(superheroes, ["Name", "Race", "Gender", "Height (m)", "Weight (kg)", "Skin color", "Hair color", "Eye color", "Alignment", "Publisher"])

superheroes.head()
```

```python
def total_nulls(df):
    return df.isnull().sum()
    
total_nulls(superheroes)
```

```python
def delete_unuseful_columns(df, percentage): #Detele columns with "100% - percentage" null values
    for column in df:
        if df[column].isnull().sum() > len(df.index)*percentage/100:
            del df[column]

delete_unuseful_columns(superheroes, 50)
    
superheroes.head()
```

```python
def rename_nulls(df): #Replace NaN with "-"
    df.fillna("-", inplace=True)
    
rename_nulls(superheroes)

superheroes.head()
```

----------SUPERHEROES FULL DATA---------------

```python
#Merge superheroes + superheroes_powers:
superheroes_full_data = pd.merge(superheroes, superheroes_powers, on='Name')
superheroes_full_data.head()
```

----------ANALYSIS---------------

```python
def column_values_sum(df, column, value): #Return column + values sum
    return column, (df[column] == value).sum()
```

```python
Powers = []
for value in superheroes_powers.columns[1:]:
    Powers.append(column_values_sum(superheroes_full_data, value, "X"))

Powers = dict(Powers)

pprint(Powers)
```

```python
plt.bar(range(len(Powers)), Powers.values(), align = "center") 
plt.xticks(range(len(Powers)), list(Powers.keys()), rotation = "vertical")
plt.title("Superheroe powers")

plt.show() 
```

```python
def column_values_sum2(df, column, value): #Return value + values sum
    return value, (df[column] == value).sum()
```

```python
Gender = []
for value in superheroes_full_data.Gender:
    Gender.append(column_values_sum2(superheroes_full_data, "Gender", value))

Gender = dict(Gender)

pprint(Gender)
```

```python
del Gender["-"] #Delete indeterminated values
```

```python
plt.bar(range(len(Gender)), Gender.values(), align = "center") 
plt.xticks(range(len(Gender)), list(Gender.keys()))
plt.title("Superheroe gender")

plt.show() 
```
