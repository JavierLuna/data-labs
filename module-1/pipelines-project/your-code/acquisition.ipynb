```python
import pandas as pd
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
def get_csv(csv_file, enconding_value):
    df = pd.read_csv(csv_file, encoding = enconding_value, na_values=["-", False])
    return df

collisions = get_csv("traffic-collision-data-from-2010-to-present.csv", "latin-1")
collisions.head(10)
```

```python
def rename_column(df, old_column, new_column):
    return df.rename(columns = {old_column: new_column}, inplace = True)

collision_type = rename_column(collisions, "Crime Code Description", "Collision type")

collisions.head(10)
```

```python
def capitalize_column_values(df, column):
    df[column] = df[column].str.capitalize()

capitalize_column_values(collisions, "Address")
capitalize_column_values(collisions, "Cross Street")
capitalize_column_values(collisions, "Premise Description")
capitalize_column_values(collisions, "Collision type")


collisions.head(9)
```

```python
collisions.drop(["LA Specific Plans", "Premise Code", "Precinct Boundaries", "Council Districts", "Victim Descent","MO Codes", "Census Tracts", "Neighborhood Councils (Certified)"], axis = 1)

collisions.head(10)
```

```python
collisions["Date Occurred"] = collisions["Date Occurred"].apply(lambda value: value[0:10])
collisions["Date Reported"] = collisions["Date Reported"].apply(lambda value: value[0:10])
            
collisions.head(10)
```

```python
collisions.sort_values(by='Date Reported', ascending=True)

collisions.head(10)
```
