```python
import pandas as pd
```

```python
cars = pd.read_csv("./data_cars/cars_train.csv")

cars.head()
```

```python
cars.corr()
```

```python
cars.columns
```

```python
def drop_column(df, column):
    df.drop(column,axis = 1, inplace = True)
```

```python
def datos_dummies(df,column):
    cars_dummies = pd.get_dummies(df,column)
    return cars_dummies
```

```python
column =['Id','county_fips', 'county_name','state_fips', 'state_code', 'state_name','city','manufacturer','odometer', 'title_status','paint_color','lat','long','transmission','make']
drop_column(cars,column)
```

```python
cars.head(5)
```

```python
cars.dtypes
```

```python
cars.dropna(inplace=True)
```

```python
column=['condition','cylinders','fuel','drive','size','type']
cars_dummies = datos_dummies(cars,column)
```

```python
X_train =cars_dummies.drop(['price'],axis = 1)
y_train = cars_dummies['price']
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import ElasticNet

regr = ElasticNet(random_state=0)
regr.fit(X_train, y_train)
```

```python
cars_test = pd.read_csv("./data_cars/cars_test.csv")

cars_test.head()
```

```python
cars_test.fillna(method='ffill',inplace=True)
```

```python
column =['Id','county_fips', 'county_name','state_fips', 'state_code', 'state_name','city','manufacturer','odometer', 'title_status','paint_color','lat','long','transmission','make']
drop_column(cars_test,column)
```

```python
column=['condition','cylinders','fuel','drive','size','type']
X_test = datos_dummies(cars_test,column)
```

```python
X_test
```

```python
y_pred = regr.predict(X_test)
len(y_pred)
```

```python
y_pred
```

```python
y_pred = pd.DataFrame(y_pred)
```

```python
cars_test = pd.read_csv("./data_cars/cars_test.csv")
cars_test = cars_test['Id']
```

```python
submi = pd.concat([cars_test, y_pred],axis = 1)
submi.columns=['Id','Price']
```

```python
submi = submi[['Id','Price']]
```

```python
submi.to_csv('submi11',index=False)
```
