```python
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
d = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
d["sickness_index"] = diabetes.target
d.head()
```

```python
d.describe()
```

```python
d.dtypes
```

```python
X = d.drop(['sickness_index', 'bp'], axis=1)
Y = d['sickness_index']
Y
# Esto es un ejemplo de aprendizaje supervisado, en el que conocemos los datos (X) y los resultados (Y).
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=5) # random_state nos permite evitar la aleatoriedad. Es como tirar un dado siempre con las mismas constantes de fuerza, humedad, etc. La ponemos a 5 por poner algo. De esta forma, la precisión siempre será la misma.
```

```python
# PRIMER MODELO: LINEAR REGRESSION

from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

# Algunos parámetros son entrenables o transformadores. fit() o transform()

regressor.fit(X_train, Y_train)
regressor.score(X_test, Y_test) # Esto nos da la precisión, la r cuadrado. Cuanto más cerca esté de 1, mejor es el modelo.
```

```python
# SEGUNDO MODELO (probamos otro modelo): RANDOM FOREST

from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(random_state=5)

regressor.fit(X_train, Y_train)
regressor.score(X_test, Y_test)
```

```python
# TERCER MODELO: NEIGHBORS

from sklearn.neighbors import KNeighborsRegressor

regressor = KNeighborsRegressor()

regressor.fit(X_train, Y_train)
regressor.score(X_test, Y_test)
```
