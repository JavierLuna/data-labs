```python
# Me cercioro de que no hay CustomerID repetidos
customer_unique = orders['CustomerID'].unique()
len(customer_unique)
```

```python
# ¿Cuantos 'amount_spent' hay por cada CustomerID?

grouped_customers = orders.groupby('CustomerID')
amount_spent = grouped_customers['amount_spent'].value_counts()
amount_spent
```

```python
# Esta es la suma total que se ha gastado cada Customer

amount_per_customer = orders.groupby("CustomerID")["amount_spent"].sum()
amount_per_customer
```
