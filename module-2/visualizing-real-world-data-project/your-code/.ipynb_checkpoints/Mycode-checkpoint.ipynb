# Huir por futuro, encontrar la muerte

```python
import plotly.plotly as py
import cufflinks as cf
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline

cf.go_offline()
```

```python
#Extraemos nuestros datos del DataFrame seleccionado:
dfemigrant = pd.read_csv('./MissingMigrants.csv', encoding='utf8')

dfemigrant.head()
```

```python
dfemigrant.apply(lambda x: x.strip(r'.!?\" \" \\n\\t') if isinstance(x, str) else x)

dfemigrant.head()
```

```python
dfemigrant['Cause of Death'].replace({'Train Accide':'Train Accident', 'Unknow':'Unknown',
                                      'Unknown (found dead next to train tracks), Train Accide': 'Train Accident', 
                                      'Train Accident, Unknown (found dead next to train tracks)': 'Train Accident',
                                      'Unknown (decomposed remains)': 'Decomposed Remains', 
                                      'Unknown (found dead next to train tracks)': 'Found dead next to train tracks',
                                      'Unknown (found dead next to train tracks), Train Accident': 'Train Accident',
                                      'Unknown (found on dinghy)': 'Found on dinghy',
                                      'Unknown (multiple blunt force injuries)': 'Multiple blunt force injuries',
                                      'Unknown (mummified and skeletal remains)': 'Mummified and skeletal remains',
                                      'Unknown (mummified remains)': 'Mummified remains',
                                      'Unknown (skeletal remains)': 'Skeletal remains', 'Unknown (violence)': 'Violence'},
                                     inplace=True)
```

```python
values = {'Number Dead': 0, 'Minimum Estimated Number of Missing':0, 'Number of Survivors': 0, 'Number of Survivors': 0, 
          'Number of Females': 0, 'Number of Males': 0, 'Number of Children': 0}
dfemigrant.fillna(value=values, inplace= True)

dfemigrant.head()
```

# Visualización de Datos

```python
pd.plotting.scatter_matrix(dfemigrant, figsize=(20,16));
```

```python
sns.violinplot("Reported Year", data=dfemigrant)
```

```python
dfemigrant.plot.scatter(x="Reported Year", y="Number Dead",c="blue")
```

```python
databar1 = dfemigrant.groupby('Region of Incident').agg({'Total Dead and Missing':'sum'}).reset_index()

databar1.iplot(kind='bar', x='Region of Incident', xTitle='Region',
               yTitle='Dead and Missing', color='orange',
               title='Total Dead and Missing')
```

```python
dfemigrant['Number Dead'].describe()
```

```python
data2 = dfemigrant.pivot_table(values='Number Dead', columns='Migration Route', 
                      index='Reported Year', aggfunc='mean')
```

```python
data2.iplot(kind='heatmap', colorscale='YlOrRd', xTitle='Year', 
           yTitle='Route', title='Avg. Mortal Routes')
```

```python
deadgender = dfemigrant.groupby('Reported Year').agg({'Number of Females':'mean','Number of Males':'mean', 'Number of Children':'mean', 'Number of Survivors':'mean' }).reset_index()

deadgender.head()
```

```python
deadgender.iplot(kind='line', x='Reported Year', xTitle= 'Year', 
           yTitle='AVG', title='Avg. mortal victims and survivors')
```

```python
dfemigrant.iplot(kind='bubble', x= 'Number Dead', y='Reported Year', size='Number Dead',
           categories='Region of Incident', xTitle='Year', yTitle='Dead',
           title='Mortal Region')
```
